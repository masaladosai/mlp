```python
'''
1. Select all the even indexed rows from the given dataset.
 What is the value in the 3rd indexed column of the 0th indexed row in the selected dataframe. (row/ column indexed from zero(0) in python/pandas)
'''
new_df=df.loc[df.index%2==0]
ques1=float(new_df.iloc[0,3])
ques1
```

```python
'''
2. Select all the even indexed rows from the given dataset.
What is the value in the 3rd indexed column of the 332th indexed row in the selected dataframe. (row/ column indexed from zero(0))
'''
ques2=float(new_df.iloc[332,3])
ques2
```

```python
'''
3. Select all the odd indexed rows from the given dataset. What is the value in the 3rd indexed column of the 332th indexed row in the selected dataframe.
(row/ column indexed from zero(0))

Please consider case sensitivity while giving your solution capital letter ≠ small letter
If it is a string don't write quotes in the solution
'''
new_df_odd=df.loc[~df.index%2==0]
ques3=float(new_df_odd.iloc[332,3])
ques3
```

```python
'''

4.
Select all the odd indexed columns from the given dataset.
 What is the value in the 5th indexed column of the 100th indexed row in the selected dataframe. (row/ column indexed from zero(0))

Please consider case sensitivity while giving your solution capital letter ≠ small letter
If it is a string don't write quotes in the solution
'''
ques4=new_df_odd.iloc[100,5]
ques4
```

```python
'''
5. Select all the even indexed columns and even indexed rows from the given dataset. What is the value in the 3th indexed column of the 255th indexed row in the selected dataframe. (row/ column indexed/starts from zero(0))

    Please consider case sensitivity while giving your solution capital letter ≠ small letter
    If it is a string don't write quotes in the solution
'''
new_df_even_even=df.iloc[::2,::2]
ques5=new_df_even_even.iloc[255,3]
ques5
```

```python
'''
6. Determine the number of samples (rows) in the dataset that belong to the six most recent years, based on the provided data.
'''
indexi=df['Year'].value_counts().index
val=list(indexi)
years_req=sorted(val)[-6:]

ques6=df['Year'][df['Year'].isin(years_req)].shape[0]
```

```python
'''
7. Determine the number of samples (rows) in the dataset which has num_rooms = 3 and num_bathrooms = 3
'''
ques7=df[(df['num_rooms']==3) & (df['num_bathrooms']==3)].shape[0]
```

```python
'''
8. Determine the number of samples (rows) in the dataset which has num_rooms = 3 or num_bathrooms = 3
'''
ques8=df[(df['num_rooms']==3) | (df['num_bathrooms']==3)].shape[0]
```

```python
'''
9. In the Year 2022, how many houses (rows) located in the Greenwich Locality have exactly num_room == 3, and facing towards either the North or East?

'''
ques9=df[(df['Year']==2022)&(df['num_rooms']==3)&(df['Face'].isin(['North','East']))].shape[0]
```

```python
'''
10. Calculate the number of samples(rows) that correspond to the month of August across all the years.?
'''
def what_month(s):
  s=s.split('-')
  return int(s[1])

new_date=df['Date'].apply(what_month).to_frame()
ques10=new_date[new_date['Date']==8].shape[0]
```

```python
'''
11. According to the data which Locality have highest average Sale Price?
'''
new_dff=df.groupby('Locality')['Sale Price'].mean()
ques11=new_dff.idxmax()
```
